Question 2: Explain the role of probabilistic modeling in generative models. How do these models differ from discriminative models?

Answer:

Probabilistic modeling plays a fundamental role in generative models because it allows these models to learn and represent the probability distribution of data. In simple terms, probabilistic modeling involves using mathematical methods to estimate how likely different data points are to occur. By learning these probability distributions, generative models can generate new data samples that resemble the training data.

In generative models, the goal is to model the joint probability distribution of the input data and its labels, usually represented as P(X, Y), where X represents the data and Y represents the label or output. By learning this joint distribution, generative models can generate new data points, simulate realistic samples, and even infer missing information.

For example, in text generation tasks, a generative model learns the probability of sequences of words appearing together. When given a prompt, the model predicts the probability of the next word and generates text by selecting words with high probabilities. Similarly, in image generation, the model learns the probability distribution of pixel patterns and can generate new images based on that learned distribution.

Probabilistic modeling also helps generative models handle uncertainty and variability in data. Instead of producing fixed outputs, the models generate results based on probability distributions, which allows them to produce diverse and creative outputs.

Difference Between Generative and Discriminative Models

Generative models and discriminative models differ mainly in what they learn and how they make predictions.

1. Generative Models

   * Generative models learn the **joint probability distribution P(X, Y).
   * They model how the data is generated and can create new data samples.
   * These models can simulate realistic data similar to the training dataset.
   * Examples include Naïve Bayes, Hidden Markov Models (HMMs), Variational Autoencoders (VAEs), and Generative Adversarial Networks (GANs).

2. Discriminative Models

   * Discriminative models learn the conditional probability distribution P(Y | X).
   * They focus only on distinguishing between different classes or predicting labels for given inputs.
   * They do not attempt to generate new data; instead, they focus on classification or regression tasks.
   * Examples include Logistic Regression, Support Vector Machines (SVMs), and Neural Network classifiers.

Key Differences

* Generative models learn how the data is generated and can create new samples, while discriminative models focus on separating or classifying existing data.
* Generative models model the full probability distribution of data, whereas discriminative models directly learn the mapping from input to output.
* Generative models are useful for tasks such as data generation, simulation, and unsupervised learning, while discriminative models are commonly used for classification and prediction tasks.

In summary, probabilistic modeling enables generative models to understand the underlying structure and distribution of data, allowing them to generate new and realistic samples. Discriminative models, on the other hand, focus primarily on making accurate predictions by learning the boundary between different classes.

Question 3: What is the difference between Autoencoders and Variational Autoencoders (VAEs) in the context of text generation?

Answer:

Autoencoders and Variational Autoencoders (VAEs) are neural network architectures used for representation learning and generative tasks. Both models learn to encode input data into a compressed representation and then reconstruct the original data from that representation. However, they differ significantly in how they represent and generate data, especially in applications such as text generation.

Autoencoders

An Autoencoder is a type of neural network designed to learn an efficient encoding of input data. It consists of two main parts: an encoder and a decoder. The encoder compresses the input data into a lower-dimensional representation called the latent vector, while the decoder reconstructs the original data from this compressed representation.

In the context of text generation, an autoencoder learns to convert a sentence or text sequence into a latent representation and then reconstruct the same sentence from it. The goal of the model is to minimize the difference between the original input and the reconstructed output.

However, traditional autoencoders are mainly designed for data compression and reconstruction**, not for generating new data. The latent space learned by a standard autoencoder may be irregular and discontinuous, which makes it difficult to sample new points from it to generate meaningful text.

Variational Autoencoders (VAEs)

Variational Autoencoders are an extension of traditional autoencoders designed specifically for generative modeling. Instead of mapping input data to a single fixed latent vector, VAEs map the input to a probability distribution in the latent space, usually represented by a mean and variance.

During training, the encoder produces parameters of this distribution, and the model samples a latent vector from it. The decoder then uses this sampled vector to generate or reconstruct the text. This probabilistic approach ensures that the latent space becomes smooth and continuous.

In text generation tasks, VAEs can generate new sentences by sampling points from the latent space and passing them through the decoder. Because the latent space follows a structured distribution (commonly a normal distribution), the model can produce diverse and meaningful outputs.

Key Differences Between Autoencoders and VAEs

1. Latent Representation

   * Autoencoders produce a single deterministic latent vector.
   * VAEs learn a probabilistic latent distribution defined by mean and variance.

2. Generative Capability

   * Autoencoders mainly reconstruct existing inputs.
   * VAEs can generate new data by sampling from the latent space.

3. Latent Space Structure

   * Autoencoders often produce an irregular latent space.
   * VAEs enforce a smooth and continuous latent space.

4. Training Objective

   * Autoencoders minimize reconstruction error.
   * VAEs optimize both reconstruction loss and a regularization term (KL divergence) to ensure the latent distribution follows a known probability distribution.

Conclusion

While both autoencoders and variational autoencoders learn compressed representations of data, VAEs are better suited for generative tasks such as text generation because they model the latent space probabilistically. This allows VAEs to generate new and diverse text samples, whereas traditional autoencoders mainly focus on reconstructing existing inputs.

Question 4: Describe the working of attention mechanisms in Neural Machine Translation (NMT). Why are they critical?

Answer:

Attention mechanisms are an important component in Neural Machine Translation (NMT) systems. They help the model focus on the most relevant parts of the source sentence when generating each word in the translated sentence. This improves translation quality, especially for long and complex sentences.

In traditional sequence-to-sequence (Seq2Seq) models used for machine translation, an encoder reads the entire source sentence and compresses it into a single fixed-length vector. The decoder then uses this vector to generate the translated sentence. However, this approach has limitations because compressing a long sentence into a single vector can lead to loss of important information.

The attention mechanism solves this problem by allowing the decoder to access all the encoder’s hidden states instead of relying on a single fixed representation.

Working of Attention Mechanism in NMT

1. Encoding the Source Sentence
   The encoder processes the input sentence word by word and produces a sequence of hidden states. Each hidden state contains information about a specific word and its context within the sentence.

2. Calculating Attention Scores
   When the decoder generates a word in the target language, it calculates a score that represents how relevant each encoder hidden state is to the current decoding step. These scores measure how much attention should be given to each word in the input sentence.

3. Generating Attention Weights
   The scores are passed through a softmax function to convert them into normalized weights (probabilities). These weights indicate the importance of each word in the source sentence for generating the next target word.

4. Creating the Context Vector
   The model computes a context vector by taking a weighted sum of all encoder hidden states using the attention weights. This context vector contains the most relevant information needed for translation at that step.

5. Producing the Output Word
   The decoder uses the context vector along with its current hidden state to generate the next word in the translated sentence.

This process repeats for each word in the target sentence, allowing the model to dynamically focus on different parts of the input sentence.

Why Attention Mechanisms Are Critical

1. Handles Long Sentences Better
   Attention allows the model to reference the entire input sequence rather than compressing it into a single vector, reducing information loss.

2. Improves Translation Accuracy
   By focusing on the most relevant words during translation, the model can produce more accurate and context-aware translations.

3. Provides Alignment Between Words
   Attention helps establish relationships between words in the source and target sentences, similar to how human translators align words across languages.

4. Enables More Advanced Models
   Attention mechanisms form the foundation of modern architectures such as the Transformer model, which relies entirely on attention mechanisms rather than recurrent networks.

Conclusion

Attention mechanisms significantly enhance Neural Machine Translation by allowing the model to dynamically focus on relevant parts of the input sentence during translation. This leads to better handling of long sentences, improved translation quality, and more efficient learning, making attention a critical component of modern NMT systems.

Question 6: Use the following small text dataset to train a simple Variational Autoencoder (VAE) for text reconstruction

Dataset:
["The sky is blue", "The sun is bright", "The grass is green", "The night is dark", "The stars are shining"]

1. The VAE Architecture
A VAE works by mapping input text into two vectors: a mean (μ) and a variance (σ). We then use the reparameterization trick to sample a latent vector z, which the decoder turns back into text.






```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# --- 1. Data Preparation ---
data = ["The sky is blue", "The sun is bright", "The grass is green",
        "The night is dark", "The stars are shining"]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(data)
sequences = tokenizer.texts_to_sequences(data)
vocab_size = len(tokenizer.word_index) + 1
max_len = max([len(seq) for seq in sequences])
X = pad_sequences(sequences, maxlen=max_len, padding='post')

# --- 2. Defining the Sampling Layer ---
class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        batch = tf.shape(z_mean)[0]
        dim = tf.shape(z_mean)[1]
        epsilon = tf.random.normal(shape=(batch, dim))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

# --- 3. Building the VAE Model Class ---
class VAE(models.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super(VAE, self).__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder
        self.total_loss_tracker = tf.keras.metrics.Mean(name="total_loss")

    def train_step(self, data):
        with tf.GradientTape() as tape:
            z_mean, z_log_var, z = self.encoder(data)
            reconstruction = self.decoder(z)

            # Reconstruction Loss
            # We use sparse_categorical_crossentropy because our targets are integers
            reco_loss = tf.reduce_mean(
                tf.reduce_sum(
                    tf.keras.losses.sparse_categorical_crossentropy(data, reconstruction),
                    axis=-1
                )
            )

            # KL Divergence Loss
            kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
            kl_loss = tf.reduce_mean(tf.reduce_sum(kl_loss, axis=1))

            total_loss = reco_loss + kl_loss

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))
        self.total_loss_tracker.update_state(total_loss)
        return {"loss": self.total_loss_tracker.result()}

# --- 4. Model Assembly ---
latent_dim = 2
intermediate_dim = 32

# Encoder
encoder_inputs = layers.Input(shape=(max_len,))
x = layers.Embedding(vocab_size, 16)(encoder_inputs)
x = layers.LSTM(intermediate_dim)(x)
z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)
z = Sampling()([z_mean, z_log_var])
encoder = models.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")

# Decoder
latent_inputs = layers.Input(shape=(latent_dim,))
x = layers.RepeatVector(max_len)(latent_inputs)
x = layers.LSTM(intermediate_dim, return_sequences=True)(x)
decoder_outputs = layers.TimeDistributed(layers.Dense(vocab_size, activation="softmax"))(x)
decoder = models.Model(latent_inputs, decoder_outputs, name="decoder")

# Instantiate and Compile
vae = VAE(encoder, decoder)
vae.compile(optimizer=tf.keras.optimizers.Adam())

# --- 5. Training and Results ---
vae.fit(X, epochs=600, verbose=0)

print("Original vs Reconstructed:")
# Get the latent representation from the encoder
_, _, z = encoder.predict(X)
# Reconstruct using the decoder
preds = decoder.predict(z)

for i in range(len(data)):
    pred_seq = np.argmax(preds[i], axis=-1)
    # Convert integer indices back to words, ignoring padding (index 0)
    reconstructed_words = [tokenizer.index_word.get(idx, "") for idx in pred_seq if idx != 0]
    print(f"Original: {data[i]}")
    print(f"VAE Result: {' '.join(reconstructed_words)}\n")
```

Question 8: Implement a simple attention-based encoder-decoder model for
English-to-Spanish translation using Tensorflow or PyTorch.

Answer:An attention-based encoder–decoder model is commonly used in neural machine translation.
The encoder reads the input English sentence and converts it into hidden representations.
The attention mechanism helps the decoder focus on relevant parts of the input sentence while generating each word.
The decoder then generates the translated Spanish sentence step by step using the context information provided by attention.

This architecture improves translation quality because the model does not rely on a single fixed context vector; instead, it dynamically attends to different parts of the input sequence during decoding.



```python
import torch
import torch.nn as nn

# Simple toy dataset
pairs = [
    ("hello", "hola"),
    ("thank you", "gracias"),
    ("good morning", "buenos dias")
]

# Vocabulary creation
input_vocab = {"<pad>":0, "hello":1, "thank":2, "you":3, "good":4, "morning":5}
output_vocab = {"<pad>":0, "hola":1, "gracias":2, "buenos":3, "dias":4}

# Encoder
class Encoder(nn.Module):
    def __init__(self, input_size, embed_size, hidden_size):
        super(Encoder, self).__init__()
        self.embedding = nn.Embedding(input_size, embed_size)
        self.rnn = nn.GRU(embed_size, hidden_size, batch_first=True)

    def forward(self, x):
        x = self.embedding(x)
        outputs, hidden = self.rnn(x)
        return outputs, hidden


# Attention mechanism
class Attention(nn.Module):
    def __init__(self, hidden_size):
        super(Attention, self).__init__()
        self.softmax = nn.Softmax(dim=1)

    def forward(self, hidden, encoder_outputs):
        scores = torch.bmm(encoder_outputs, hidden.permute(1,2,0))
        attn_weights = self.softmax(scores)
        context = torch.bmm(attn_weights.permute(0,2,1), encoder_outputs)
        return context


# Decoder
class Decoder(nn.Module):
    def __init__(self, output_size, embed_size, hidden_size):
        super(Decoder, self).__init__()
        self.embedding = nn.Embedding(output_size, embed_size)
        self.rnn = nn.GRU(embed_size + hidden_size, hidden_size, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x, hidden, context):
        x = self.embedding(x)
        rnn_input = torch.cat((x, context), dim=2)
        output, hidden = self.rnn(rnn_input, hidden)
        prediction = self.fc(output)
        return prediction, hidden


# Model parameters
embed_size = 8
hidden_size = 16

encoder = Encoder(len(input_vocab), embed_size, hidden_size)
attention = Attention(hidden_size)
decoder = Decoder(len(output_vocab), embed_size, hidden_size)

print("Attention-based Encoder-Decoder model initialized successfully.")
```
